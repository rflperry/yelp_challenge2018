```python
df = pd.read_csv('./dataset/H113.csv', encoding='utf-8')
wiki_page_names = np.array(df.WikiPageName)
```

```python
class Politician:
    def __init__(self, d):
        self.WikiPageName = str(d['WikiPageName'])
        self.Party = str(d['Party'])
        self.State = str(d['State'])
    def __hash__(self):
        return hash(self.WikiPageName)
    def __eq__(self, other):
        return (
                self.__class__ == other.__class__ and 
                self.WikiPageName == other.WikiPageName
               )
```

```python
def get_article_links(name, year):
    article_pattern = r'\[\[([^\]]*?)(?:\|.*?)*\]\]'
    path = './congress2015/%s%s.txt' % (name, year)
    article = io.open(path, 'r', encoding='utf-8').read()
    article_links = re.findall(article_pattern, article)
    article_links = [a.replace(' ', '_') for a in article_links]
    return article_links
```

```python
DG = nx.DiGraph()

for i,name in enumerate(wiki_page_names):
    article_links = get_article_links(name, "2015")
    p = Politician(dict(df.iloc[i]))
    DG.add_node(p)
    for link in article_links:
        if link in wiki_page_names:
            j = wiki_page_names == link
            q = Politician(dict(df.iloc[j]))
            DG.add_edge(p, q)
```

```python
print( 'Nodes', len(DG.nodes()) )
print( 'Links', len(DG.edges()) ) 
```

```python
degrees = DG.degree()
```
